```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install",
                       "--upgrade", "scikit-learn", "-q"])

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.utils import resample
import xgboost as xgb
import joblib

# Set global random seed for reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
# Load each Parquet file
df1 = pd.read_parquet("/home/elidrissi/mlDataset/DoS1-Thursday-15-02-2018_TrafficForML_CICFlowMeter.parquet")
df2 = pd.read_parquet("/home/elidrissi/mlDataset/DoS2-Friday-16-02-2018_TrafficForML_CICFlowMeter.parquet")
df3 = pd.read_parquet("/home/elidrissi/mlDataset/DDoS1-Tuesday-20-02-2018_TrafficForML_CICFlowMeter.parquet")
df4 = pd.read_parquet("/home/elidrissi/mlDataset/DDoS2-Wednesday-21-02-2018_TrafficForML_CICFlowMeter.parquet")

data = pd.concat([df1, df2, df3, df4], ignore_index=True)
print(f"Total rows loaded: {len(data)}")
print(f"Columns: {data.columns.tolist()}")
```

```python
data = data[data['Label'] != 'DoS attacks-SlowHTTPTest']
```

```python
data = data[data['Label'] != 'DDOS attack-LOIC-UDP']
```

```python
print(data['Label'].value_counts())
```

```python
# Keep only the 6 classes 
keep_labels = [
    'Benign',
    'DoS attacks-GoldenEye',
    'DoS attacks-Slowloris',
    'DoS attacks-Hulk',
    'DDoS attacks-LOIC-HTTP',
    'DDOS attack-HOIC'
]

data = data[data['Label'].isin(keep_labels)].copy()
print("\nClass counts after filtering:")
print(data['Label'].value_counts())
```

```python
# Identify numeric columns (exclude 'Label')
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
zero_pct = (data[numeric_cols] == 0).mean() * 100

# Columns that are 100% zeros -> drop them (no information)
cols_100_zero = zero_pct[zero_pct == 100].index.tolist()
print(f"Columns with 100% zeros: {cols_100_zero}")
data.drop(columns=cols_100_zero, inplace=True)

# Handle possible infinite values (replace with 0)
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(0, inplace=True)

print(f"Remaining columns after dropping 100% zeros: {data.shape[1]}")
```

```python
le = LabelEncoder()
data['Label_encoded'] = le.fit_transform(data['Label'])
class_names = le.classes_
print("Encoded classes:", class_names)
```

```python
# Separate features and target
X = data.drop(columns=['Label', 'Label_encoded'])
y = data['Label_encoded']

# Split (stratified to preserve class proportions)
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print(f"Training set size: {len(X_train_full)}")
print(f"Test set size: {len(X_test)}")
```

```python
print(data['Label'].unique())

dos_ddos_labels = [
    'Benign',
    'DoS attacks-GoldenEye',
    'DoS attacks-Slowloris',
    'DoS attacks-Hulk',
 
    'DDOS attack-HOIC'   ]    

df = data[data['Label'].isin(dos_ddos_labels )]
print(df['Label'].value_counts())
```

```python
# Determine majority class and its count in the training set
train_counts = y_train_full.value_counts()
majority_label = train_counts.idxmax()
majority_count = train_counts.max()
second_count = train_counts.sort_values(ascending=False).iloc[1]
target_majority = second_count  # match the second largest class

print(f"Majority class: {class_names[majority_label]} (count: {majority_count})")
print(f"Target count after downsampling: {target_majority}")

# Create a temporary DataFrame with features + target
train_df = X_train_full.copy()
train_df['Label_encoded'] = y_train_full

# Split majority and minority
majority_df = train_df[train_df['Label_encoded'] == majority_label]
minority_df = train_df[train_df['Label_encoded'] != majority_label]

# Downsample majority
downsampled_majority = resample(majority_df,
                                replace=False,
                                n_samples=target_majority,
                                random_state=RANDOM_STATE)

balanced_df = pd.concat([downsampled_majority, minority_df], ignore_index=True)
balanced_df = balanced_df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

print("\nClass distribution after downsampling:")
print(balanced_df['Label_encoded'].value_counts().sort_index())

# Separate features and target for the balanced training set
X_train = balanced_df.drop(columns=['Label_encoded'])
y_train = balanced_df['Label_encoded']

# Compute sample weights: majority gets downsampling factor, others get 1
from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)
```

```python
# ✅ Code corrigé
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)   # fit sur train uniquement
X_test_scaled  = scaler.transform(X_test)         # transform sur test



full_feature_columns = X_train.columns.tolist()  # ← ADD THIS LINE
k_features = 60
selector = SelectKBest(mutual_info_classif, k=k_features)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected  = selector.transform(X_test_scaled)

selected_features = X_train.columns[selector.get_support()].tolist()
print(f"Selected {len(selected_features)} features")

X_train = pd.DataFrame(X_train_selected, columns=selected_features)
X_test  = pd.DataFrame(X_test_selected,  columns=selected_features)
```

```python
# Split training data into train and validation (stratified)
X_train_sub, X_val, y_train_sub, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y_train
)

# Corresponding sample weights
# ✅ Code corrigé
X_train_sub, X_val, y_train_sub, y_val, sample_weights_sub, sample_weights_val = train_test_split(
    X_train, y_train, sample_weights,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y_train
)
```

```python
#test pos_weight

def get_custom_sample_weights(y, class_weights_dict):
    """
    Given an array of labels (y) and a dict mapping class_idx -> weight,
    return an array of sample weights.
    """
    sample_weights = np.ones(len(y))
    for class_idx, weight in class_weights_dict.items():
        sample_weights[y == class_idx] = weight
    return sample_weights




custom_weights_B = get_custom_sample_weights(y_train, {
    0: 0.8,    # Slightly lower for Benign
    1: 2.0,
    2: 1.5,
    3: 5.0,
    4: 2.5,
    5: 8.0
})    
```

```python
# After the train_test_split that creates X_train_sub, X_val, etc.
# (you already have these variables from Cell 26)

# Subset custom_weights_B for train and validation
custom_weights_B_sub = custom_weights_B[y_train.index.isin(X_train_sub.index)]
custom_weights_B_val = custom_weights_B[y_train.index.isin(X_val.index)]





xgb_model = xgb.XGBClassifier(
    n_estimators=1000,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    tree_method='hist',
    device='cuda',                     # use 'cpu' if no GPU
    eval_metric='mlogloss',
    early_stopping_rounds=30,          # <--- MOVED HERE
    random_state=RANDOM_STATE,
    n_jobs=-1
)

# Train with early stopping – eval_set goes in fit()
xgb_model.fit(
    X_train_sub, y_train_sub,
    sample_weight=custom_weights_B_sub,          # <-- use subset
    eval_set=[(X_val, y_val)],
    sample_weight_eval_set=[custom_weights_B_val], # <-- use subset
    verbose=50
)

print(f"\nBest iteration: {xgb_model.best_iteration}")
```

```python
# Add this after training:
y_pred_train = xgb_model.predict(X_train_sub)
y_pred_test  = xgb_model.predict(X_test)

train_f1 = f1_score(y_train_sub, y_pred_train, average='weighted')
test_f1  = f1_score(y_test,      y_pred_test,  average='weighted')

print(f"Train F1 : {train_f1:.4f}")
print(f"Test  F1 : {test_f1:.4f}")
print(f"Gap      : {abs(train_f1 - test_f1):.4f}  {'✅ OK' if abs(train_f1-test_f1) < 0.01 else '⚠️ Overfitting!'}")
```

```python
#  Code corrigé — optimisation globale sur F1 weighted
from sklearn.calibration import CalibratedClassifierCV

def find_optimal_thresholds(model, X_val, y_val, le):
    """
    Optimise les seuils par classe en maintenant la cohérence
    globale : on choisit toujours la classe avec la plus haute
    probabilité parmi celles qui dépassent leur seuil.
    """
    y_proba = model.predict_proba(X_val)
    n_classes = len(le.classes_)
    thresholds = {i: 0.5 for i in range(n_classes)}

    print("=" * 50)
    print("OPTIMIZING PER-CLASS THRESHOLDS (global F1)")
    print("=" * 50)

    best_global_f1 = 0.0
    best_thresholds = thresholds.copy()

    # Grille de recherche sur le seuil de Benign uniquement
    # (les autres classes gardent 0.5 comme référence)
    for benign_thresh in np.arange(0.30, 0.95, 0.05):
        candidate = {i: 0.5 for i in range(n_classes)}
        candidate[0] = benign_thresh  # classe Benign

        # Simuler les prédictions avec ces seuils
        y_pred_custom = []
        for probs in y_proba:
            passed = {i: probs[i] for i in range(n_classes) if probs[i] >= candidate[i]}
            if passed:
                pred = max(passed, key=passed.get)
            else:
                pred = int(np.argmax(probs))
            y_pred_custom.append(pred)

        f1 = f1_score(y_val, y_pred_custom, average='weighted', zero_division=0)
        if f1 > best_global_f1:
            best_global_f1 = f1
            best_thresholds = candidate.copy()

    for idx, name in enumerate(le.classes_):
        print(f"  Class {idx} ({name}): threshold = {best_thresholds[idx]:.2f}")
    print(f"  → Best global F1 (weighted): {best_global_f1:.4f}")
    print("=" * 50)
    return best_thresholds
```

```python
optimal_thresholds = find_optimal_thresholds(xgb_model, X_val, y_val, le)
```

```python
# ✅ Appliquer les seuils customisés
def predict_with_thresholds(model, X, thresholds, n_classes):
    y_proba = model.predict_proba(X)
    y_pred = []
    for probs in y_proba:
        passed = {i: probs[i] for i in range(n_classes) if probs[i] >= thresholds[i]}
        pred = max(passed, key=passed.get) if passed else int(np.argmax(probs))
        y_pred.append(pred)
    return np.array(y_pred)

y_pred = predict_with_thresholds(xgb_model, X_test, optimal_thresholds, len(le.classes_))
y_pred_proba = xgb_model.predict_proba(X_test)

# Overall metrics
accuracy = accuracy_score(y_test, y_pred)
f1_weighted = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print("="*60)
print("FINAL EVALUATION (80/20 random stratified split)")
print("="*60)
print(f"Accuracy  : {accuracy:.4f}")
print(f"F1 (weighted) : {f1_weighted:.4f}")
print(f"Precision (weighted): {precision:.4f}")
print(f"Recall (weighted)   : {recall:.4f}")
print("="*60)

# Per-class report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix – 80/20 Split')
plt.tight_layout()
plt.show()
```

```python
def predict_blind_csv_row_custom(model, df_csv, full_feature_columns, 
                                  thresholds, le, scaler, selector):
    # Input validation
    while True:
        try:
            user_input = input(f"\n🔹 Enter a 2017 row index (0 to {len(df_csv)-1}): ")
            row_idx = int(user_input)
            if 0 <= row_idx < len(df_csv):
                break
            else:
                print(f"❌ Out of range!")
        except ValueError:
            print("❌ Invalid input.")

    # Extract and align row
    raw_row = df_csv.iloc[row_idx].to_dict()
    norm_2017 = {}
    for col, val in raw_row.items():
        base = col.split('.')[0]
        norm_2017[normalize_string(base)] = val

    aligned_sample = {}
    for model_col in full_feature_columns:
        norm_model = normalize_string(model_col)
        if norm_model in norm_2017:
            aligned_sample[model_col] = norm_2017[norm_model]
        elif norm_model in alias_dict:
            matched = False
            for alias in alias_dict[norm_model]:
                if alias in norm_2017:
                    aligned_sample[model_col] = norm_2017[alias]
                    matched = True
                    break
            if not matched:
                aligned_sample[model_col] = 0.0
        else:
            aligned_sample[model_col] = 0.0

    # Clean values
    for col in aligned_sample:
        try:
            val = float(aligned_sample[col])
            aligned_sample[col] = 0.0 if (np.isnan(val) or np.isinf(val)) else val
        except:
            aligned_sample[col] = 0.0

    # Build raw DataFrame (all original features before selection)
    input_raw = pd.DataFrame([aligned_sample])[full_feature_columns]

    # ✅ Step 1 — Apply the SAME scaler used during training
    input_scaled = pd.DataFrame(
        scaler.transform(input_raw),
        columns=full_feature_columns
    )

    # ✅ Step 2 — Apply the SAME feature selector used during training
    input_selected = pd.DataFrame(
        selector.transform(input_scaled),
        columns=selected_features
    )

    # Predict
    pred_probs = model.predict_proba(input_selected)[0]

    # Apply thresholds
    best_class = 0
    best_score = 0.0
    for class_idx, thresh in thresholds.items():
        prob = pred_probs[class_idx]
        if prob >= thresh and prob > best_score:
            best_score = prob
            best_class = class_idx

    confidence = pred_probs[best_class] * 100

    # Print report
    print("\n" + "="*50)
    print("🛡️  PREDICTION (2017 DATA — FIXED PIPELINE)")
    print("="*50)
    print(f" 🔹 Row Index     : {row_idx}")
    print(f" 🔹 Decision      : {le.classes_[best_class]} (encoded: {best_class})")
    print(f" 🔹 Confidence    : {confidence:.2f}%")
    print("="*50)
    if best_class == 0:
        print(" 🟩 STATUS: TRAFFIC SEEMS CLEAR / SAFE")
    else:
        print(" 🚨 STATUS: MALICIOUS ACTIVITY FLAGGED!")
    print("-"*50)
    print(" 📊 Probability Distribution:")
    for idx, name in enumerate(le.classes_):
        thresh = thresholds.get(idx, 0.5)
        passed = "✅" if pred_probs[idx] >= thresh else "⬜"
        print(f"   • {str(name):<26}: {pred_probs[idx]*100:>6.2f}%  "
              f"(threshold: {thresh:.2f}) {passed}")
    print("="*50 + "\n")


# ✅ Call with scaler and selector passed in
predict_blind_csv_row_custom(
    xgb_model, df_2017, full_feature_columns,
    simple_thresholds, le,
    scaler,     # ← pass your fitted scaler
    selector    # ← pass your fitted selector
)
```

```python
import pandas as pd
import numpy as np

# Step 1 — Check which selected features are MISSING in 2017
print("=" * 55)
print("FEATURE COVERAGE ANALYSIS")
print("=" * 55)

# Normalize 2017 column names
df_2017_norm = {normalize_string(c): c for c in df_2017.columns}

missing_in_2017 = []
found_in_2017   = []

for feat in selected_features:
    norm = normalize_string(feat)
    if norm in df_2017_norm or norm in alias_dict:
        found_in_2017.append(feat)
    else:
        missing_in_2017.append(feat)

print(f"\n✅ Found in 2017  : {len(found_in_2017)}/{len(selected_features)}")
print(f"❌ Missing in 2017: {len(missing_in_2017)}/{len(selected_features)}")
print(f"\nMissing features:\n{missing_in_2017}")
```

```python
# Step 2 — Check feature distribution shift
print("\n" + "=" * 55)
print("DISTRIBUTION SHIFT ANALYSIS (top 10 features)")
print("=" * 55)

for feat in selected_features[:10]:
    norm = normalize_string(feat)
    if norm in df_2017_norm:
        col_2017 = df_2017_norm[norm]
        mean_2018 = X_train[feat].mean()
        mean_2017 = pd.to_numeric(df_2017[col_2017], errors='coerce').mean()
        ratio = mean_2017 / (mean_2018 + 1e-9)
        flag = "⚠️" if ratio > 2 or ratio < 0.5 else "✅"
        print(f"{flag} {feat:<35} 2018: {mean_2018:>12.2f}  2017: {mean_2017:>12.2f}")
```

```python
print("=" * 55)
print("DISTRIBUTION SHIFT ANALYSIS (RAW DATA)")
print("=" * 55)

# ✅ Use X_train_full (raw, before scaling)
for feat in selected_features[:10]:
    norm = normalize_string(feat)
    
    # Get 2018 raw mean
    mean_2018 = X_train_full[feat].mean() if feat in X_train_full.columns else None
    
    # Get 2017 raw mean via alias
    col_2017 = None
    if norm in df_2017_norm:
        col_2017 = df_2017_norm[norm]
    elif norm in alias_dict:
        for alias in alias_dict[norm]:
            if alias in df_2017_norm.values():
                col_2017 = alias
                break

    if mean_2018 is not None and col_2017 is not None:
        mean_2017 = pd.to_numeric(df_2017[col_2017], errors='coerce').mean()
        ratio = mean_2017 / (mean_2018 + 1e-9)
        flag = "⚠️" if ratio > 3 or ratio < 0.33 else "✅"
        print(f"{flag} {feat:<35} 2018: {mean_2018:>14.2f}  "
              f"2017: {mean_2017:>14.2f}  ratio: {ratio:.2f}x")
    else:
        print(f"❌ {feat:<35} → not found in 2017")
```
