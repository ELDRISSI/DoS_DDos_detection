```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing & Data Utilities
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Model Execution
from sklearn.ensemble import RandomForestClassifier

# Comprehensive Evaluation Metrics
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix, classification_report, 
                             roc_curve, auc)

print("✅ Cell 1 Complete: All libraries imported successfully.")
```

```python
# Define path to the shared datasets folder (one level up from /random_forest)
DATA_DIR = "../datasets"

# Automatically find all Parquet files in that folder
files = [f for f in os.listdir(DATA_DIR) if f.endswith('.parquet')]
print(f"🔍 Found files: {files}")

# Read and merge all files into a single master DataFrame
dfs = [pd.read_parquet(os.path.join(DATA_DIR, f)) for f in files]
df_full = pd.concat(dfs, ignore_index=True)

print(f"\n📦 Cell 2 Complete: Combined dataset loaded!")
print(f"Total Rows: {df_full.shape[0]:,}")
print(f"Total Columns (Features): {df_full.shape[1]}")
```
